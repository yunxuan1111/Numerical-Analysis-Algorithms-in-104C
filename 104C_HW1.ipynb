#### Problem 3 Conjugate Gradient Method

```python
import numpy as np
from numpy import linalg as LA
```

```python
def cg(A,b,x):
    r = b - np.matmul(A,x)
    v = r
    while LA.norm(r) > 10**(-6):
        t = (np.matmul(np.transpose(r),r))/(np.matmul(np.transpose(v),np.matmul(A,v)))
        x = x + t*v
        r1 = r - t*np.matmul(A,v)
        s = (np.matmul(np.transpose(r1),r1))/(np.matmul(np.transpose(r),r))
        v = r1 + s*v
        r = r1
    return x
        
A = np.array([[10,1,2,3,4],[1,9,-1,2,-3],[2,-1,7,3,-5],[3,2,3,12,-1],[4,-3,-5,-1,15]])
b = np.array([[12],[-27],[14],[-17],[12]])
x = np.zeros((5,1))
cg(A,b,x) #Print the solution for x
```

#### Problem 4 Power Method

```python
def power(A,v):
    k = 0
    r = np.full((10, 1), 100)
    while LA.norm(r) > 10**(-6):
        v = np.matmul(A,v)
        v = v / LA.norm(v)
        lam = np.matmul(np.matmul(np.transpose(v),A),v)
        r = np.matmul(A,v) - lam*v
        k = k+1
    return lam,v
    

c = (-1)*np.ones(9)     # subdiagonal
a = 2*np.ones(10)   # main diagonal
b = (-1)*np.ones(9)    # superdiagonal
# Combine the superdiagonal, main diagonal, and subdiagonal to form the tridiagonal matrix A and solve the problem
A = np.diag(a)+np.diag(b,1)+np.diag(c,-1)
v = np.random.rand(10,1)
power(A,v) # run the function to get the eigenvalue and eigenvector
```

#### Problem 5 Newton's Method

```python
f = lambda x: x - np.tan(x)       # define the function f(x) = x - tanx
fp = lambda x: -np.tan(x)**2      # f'(x) = -(tanx)^2


def newton(x):
    for i in range(1000):
        x = x - (f(x))/(fp(x))
        if abs(x) < 10**(-6):
            break
    return x

print(newton(4.5))    # the root nearest 4.5
print(newton(7.7))    # the root nearest 7.7
```
